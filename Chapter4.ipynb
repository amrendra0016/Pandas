```python
# covert the Snow data into Cold in event column and windspeed 10 to 11 in Windspeed column.
# Learn to use converters
def convert_windspeed(cell):
    if cell==10:
        return 11
    return cell

def convert_event(cell):
    if cell=='Snow':
        return 'Cold'
    return cell
df9=pd.read_excel("/Users/amrendraupadhyay/Downloads/weather.xlsx","Sheet1",converters={
    'Windspeed':convert_windspeed,
    'event':convert_event
})
df9
```

```python
# save the dataframe by offsetting 2 rows down and 1 coloumn from left
df9.to_excel("/Users/amrendraupadhyay/Downloads/new.xlsx", sheet_name="data1", index=False, startrow=2, startcol=1)
```

## Write down Two dataframe to Two different sheet

```python
df_stocks = pd.DataFrame({
    'tickers': ['GOOGL', 'WMT', 'MSFT'],
    'price': [845, 65, 64 ],
    'pe': [30.37, 14.26, 30.97],
    'eps': [27.82, 4.61, 2.12]
})

df_weather =  pd.DataFrame({
    'day': ['1/1/2017','1/2/2017','1/3/2017'],
    'temperature': [32,35,28],
    'event': ['Rain', 'Sunny', 'Snow']
})
```

```python
with pd.ExcelWriter('/Users/amrendraupadhyay/Downloads/stocks_weather.xlsx') as writer:
    df_stocks.to_excel(writer, sheet_name="stocks")
    df_weather.to_excel(writer, sheet_name="weather")
```
