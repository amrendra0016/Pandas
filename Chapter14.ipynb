# Resampling

```python
# resampling dates to monthly
df['Close'].resample('M').mean().head()
```

```python
# plotting the close data, since here index is date so x axis will show dates
%matplotlib inline
df['Close'].plot()
```

```python
# Plotting the graph in a bar form

df['Close'].resample('M').mean().plot(kind='bar')
```
