```python
def get_dom(dt):
    return dt.day

df['sdom'] = df['Start Time'].map(get_dom)
```

```python
def get_dom(dt):
    return dt.month

df['s_month'] = df['Start Time'].map(get_dom)
```

```python
df
```

```python
df['month'] = df['Start Time'].dt.month_name()
df
```

# display the most common month


```python
groupby_df=df.groupby("month")
groupby_df
```

```python
#l=[]
d={}
for keys,values in groupby_df:
    #l.append(values.count().max())
    d[keys]=values.count().max()
    print(keys,values.count().max())
d
```

```python
# finding key having maximum value 
max(d,key=d.get)
```
