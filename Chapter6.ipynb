```python
import numpy as np
import pandas as pd
df = pd.read_csv("/Users/amrendraupadhyay/Downloads/weather_data.csv")
df
```

# Using replace function


```python
new_df=df.replace(-99999,np.NaN)
new_df
```

```python
# we can pass list of values to be replaced
new_df2=df.replace([-99999,'No Event'],np.NaN)
new_df2
```

```python
# we can pass dictionary also
new_df3=df.replace({
    'Temp':-99999,
    'W-Speed':-99999,
    'Event':'No Event'
},np.NaN)
new_df3
```

```python
# Replacing by using mapping
new_df4=df.replace({
    -99999: np.NaN,
    "No Event" : 'Snow'
})
new_df4
```
