```python
# parse_dates is used to change the str type to date type while reading csv file
df=pd.read_csv("/Users/amrendraupadhyay/Downloads/new.csv",parse_dates=["Day"])
type(df.Day[0])
```

```python
# method=ffill fills the na value with previous row value
df.fillna(method="ffill")
```

```python
# method=bfill fills the na value by copying next row value.
df.fillna(method="bfill")
```

```python
# by using axis="column we have copied column data here we have copied next column data to previous column data"
df.fillna(method="ffill", axis="columns")
```

```python
newdf=pd.read_csv("/Users/amrendraupadhyay/Downloads/new.csv",parse_dates=["Day"])
newdf
```

```python
# when we want to fill NaN but in limited no.
new1_df = newdf.fillna(method="ffill",limit=1)
new1_df
```

```python
# When we want to insert values but more logically
new1_df = newdf.interpolate()
new1_df
```

```python
# when we want to drop all rows containing Single or multiple NaN value
new1_df = newdf.dropna()
new1_df
```

```python
newdf
```

```python
# drop rows in which all values are NaN
new1_df = newdf.dropna(how='all')
new1_df
```

```python
# thresh =1 means if we have atleast one non NaN value than keep the row otherwise drop
new1_df = newdf.dropna(thresh=1)
new1_df
```

```python
new1_df = newdf.dropna(thresh=2) # i need atlest two valid values
new1_df
```

```python
new1_df.set_index('Day',inplace=True)
```
