Now using pivot to transform rows into columns
Syntax :- DataFrame.pivot(self, index=None, columns=None, values=None) â†’ 'DataFrame'

```python
df.pivot(index='day',columns='city')
```

```python
# to know a particular values
df.pivot(index='day',columns='city',values='temperature')
```

```python
# we can chage the values to see the different dataframes
df.pivot(index='city',columns='day')
```

Margins :- margins are used to calculate total  values both rows wise and column wise


```python
# By default aggregate function will find mean
df.pivot_table(index="city",columns="date", margins=True)
```

```python
df.pivot_table(index="city",columns="date", margins=True,aggfunc=np.sum)
```

```python
df.pivot_table(index="city",columns="date",aggfunc=np.sum)
```

# Grouper

```python
df=pd.read_csv("new.csv")
df
```

```python
# changing the data type of date column to date earlier it was string
df['date']=pd.to_datetime(df['date'])
```

```python
# Grouper groups the data based on key given and frequency provided.
df.pivot_table(index=pd.Grouper(freq='M',key='date'),columns='city')
```

```python
# Grouping on daily basis
df.pivot_table(index=pd.Grouper(freq='D',key='date'),columns='city')
```

```python
# Grouping on weekly basis
df.pivot_table(index=pd.Grouper(freq='W',key='date'),columns='city')
```
