# Creating a data frame using key value dictionary

```python
weather_data_dict = {
    'day': ['1/1/2017','1/2/2017','1/3/2017'],
    'temperature': [32,35,28],
    'windspeed': [6,7,2],
    'event': ['Rain', 'Sunny', 'Snow']
}
df2 = pd.DataFrame(weather_data_dict)
df2
```

# Creating a data frame using List of tuples


```python
weather_data_list_of_tuple = [
    ('1/1/2017',32,6,'Rain'),
    ('1/2/2017',35,7,'Sunny'),
    ('1/3/2017',28,2,'Snow')
]
df3 = pd.DataFrame(data=weather_data_list_of_tuple, columns=['day','temperature','windspeed','event'])
df3
```
